# **FUNCTION REMOVE STOPWORDS AND CLEANING OF DATA**

```python
def remove_stopwords(data_list):
    our_stopwords = { 'does', 'explain','between','which','to', 'is','what','a', 'and','are','an','the','he','she','that','?','in','or','can','be','will','you','of','at','when','how','do','mean','by','why','her','he','it','am'}
    store = []
    for i in range(len(data_list)):
        data = str(data_list[i])
        data = re.sub('[^a-zA-Z]', ' ',data)  # removing all expresions except alphabets
        data = data.lower()
        data = data.split() #converts to list of words
        # store1.append(data)
        data = [word for word in data if word not in our_stopwords] #list comprehension
        data = ' '.join(data)
        store.append(data)
    return store
```

# **PAST 4 YEAR WISE ANALYSIS**

```python
def Sort(sub_li):
    sub_li.sort(key = lambda x: x[1])
    return sub_li
def year_wise(data_list,year):
  result = []
  cn_count = 0
  os_count = 0
  dbms_count = 0
  cpp_count = 0
  java_count = 0
  other_count = 0
  k = 0
  for lists in data_list:
    k+=1
    if k>len(data_list) : break
    for word in lists :
      temp_cn = 0
      temp_os = 0
      temp_dbms = 0
      temp_cpp = 0
      temp_java = 0
      cn_topic_count = set()
      os_topic_count = set()
      dbms_topic_count = set()
      cpp_topic_count =  set()
      java_topic_count = set()
      cn_flag = False
      os_flag = False
      dbms_flag = False
      cpp_flag = False
      java_flag = False

      for i in cn:
        temp = i[1]
        if(temp.count(word)):
          cn_topic_count.add(i[0])
          temp_cn+=1
      for i in os:
        temp = i[1]
        if(temp.count(word)):
          os_topic_count.add(i[0])
          temp_os+=1
      for i in dbms:
        temp = i[1]
        if(temp.count(word)):
          dbms_topic_count.add(i[0])
          temp_dbms+=1
      for i in cpp:
        temp = i[1]
        if(temp.count(word)):
          cpp_topic_count.add(i[0])
          temp_cpp+=1
      for i in java:
        temp = i[1]
        if(temp.count(word)):
          java_topic_count.add(i[0])
          temp_java+=1
    max_box = []
    max_box.append(['cn',temp_cn])
    max_box.append(['os',temp_os])
    max_box.append(['dbms',temp_dbms])
    max_box.append(['cpp',temp_cpp])
    max_box.append(['java',temp_java])
    ans = Sort(max_box)

    if ans[4][0] == 'cn' and ans[4][1]>=1 :
      cn_count+=1
    elif ans[4][0] == 'os' and ans[4][1]>=1 :
      os_count+=1
    elif ans[4][0] == 'dbms' and ans[4][1]>=1 :
      dbms_count+=1
    elif ans[4][0] == 'cpp' and ans[4][1]>=1 :
      cpp_count+=1
    elif ans[4][0] == 'java' and ans[4][1]>=1 :
      java_count+=1
    else:
        other_count+=1
  print(f"<<<<<<<<<< Percentage of Questions in Year {year} >>>>>>>>>>")
  print("CN Questions : ",cn_count/len(data_list) *100)
  print("OS Questions : ",os_count/len(data_list) *100)
  print("DBMS Questions : ",dbms_count/len(data_list) *100)
  print("CPP Questions : ",cpp_count/len(data_list) *100)
  print("JAVA Questions : ",java_count/len(data_list) *100)
  print("MIX TYPE Questions : ",other_count/len(data_list) *100)
  result = [cn_count,os_count,dbms_count,cpp_count,java_count]
  return result
    
```

```python
result_2019 = year_wise(ques_2019,2019)
print("-------------------------------------------------------------------")
result_2020 = year_wise(ques_2020,2020)
print("-------------------------------------------------------------------")
result_2021 = year_wise(ques_2021,2021)
print("-------------------------------------------------------------------")
result_2022 = year_wise(ques_2022,2022)
print("-------------------------------------------------------------------")
```
